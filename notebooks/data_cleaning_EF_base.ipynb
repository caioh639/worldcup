```python
fase_map = {
    'group stage': 'Fase de Grupos',
    'round of 16': 'Oitavas de Final',
    'quarter-finals': 'Quartas de Final',
    'semi-finals': 'Semi-finais',
    'final': 'Final',
    'third-place match': 'Terceiro Lugar',
    'first round': 'Fase de Grupos',
    'first group stage': 'Fase de Grupos',
    'second round': 'Fase de Grupos',
    'second group stage': 'Fase de Grupos',
    'group stage play-off': 'Fase de Grupos',
    'final stage': 'Fase de Grupos'}
```

```python
matches ['Stage'] =  (matches['Round'].str.lower().str.strip().map(fase_map))

#Criando uma nova coluna 'Stage' mapeando os valores da coluna 'Round' para categorias mais amplas, 
# utilizando um dicionário de mapeamento para agrupar as fases do torneio em categorias como 'Fase de Grupos', 
# 'Oitavas de Final', etc.
```

```python
def get_winner(row):
    if pd.notna(row['home_penalty']) and pd.notna(row['away_penalty']):
        if row['home_penalty'] > row['away_penalty']:
            return row['home_team']
        elif row['home_penalty'] < row['away_penalty']:
            return row['away_team']
        else:
            return 'Draw'
    elif row['home_score'] > row['away_score']:
        return row['home_team']
    elif row['home_score'] < row['away_score']:
        return row['away_team']
    else:
        return 'Draw'
    

#Função para determinar o vencedor de cada partida, considerando tanto o resultado do tempo normal quanto os pênaltis, e retornando o nome do time vencedor ou 'Draw' em caso de empate.
```

```python
matches ['Winner'] = matches.apply(get_winner, axis=1  )
```

```python
def count_cards(row):
    if pd.isna(row):
        0
    try: 
        eventos = ast.literal_eval(str(row))
        return len(eventos)
    except: 
        return 0
    
#Função para contar o número de cartões em uma célula, 
# tratando valores ausentes e convertendo strings que representam listas de eventos em objetos Python 
# para contar corretamente o número de cartões.
```

```python
def count_red_cards(x):
    if pd.isna(x) or str(x).strip() == '':
        return 0
    
    return str(x).count('|') + 1
```

```python
matches ['h_yellow_card_long_count'] = matches['home_yellow_card_long'].apply(count_cards)

matches ['a_yellow_card_long_count'] = matches['away_yellow_card_long'].apply(count_cards)
```

```python
matches['h_red_card_count'] = (
    matches['home_red_card']
    .apply(count_red_cards)
)

matches['a_red_card_count'] = (
    matches['away_red_card']
    .apply(count_red_cards)
)
```

```python
matches['h_yellow_red_card_count'] = (
    matches['home_yellow_red_card']
    .apply(count_red_cards)
)

matches['a_yellow_red_card_count'] = (
    matches['away_yellow_red_card']
    .apply(count_red_cards)
)
```
