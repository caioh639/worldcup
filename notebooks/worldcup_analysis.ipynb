```python
matches ['Date'] = pd.to_datetime(matches['Date'], format = '%Y-%m-%d', errors = 'coerce')
```

```python
print(matches['Round'].value_counts())
```

```python
fase_map = {
    'group stage': 'Fase de Grupos',
    'round of 16': 'Oitavas de Final',
    'quarter-finals': 'Quartas de Final',
    'semi-finals': 'Semi-finais',
    'final': 'Final',
    'third-place match': 'Terceiro Lugar',
    'first round': 'Fase de Grupos',
    'first group stage': 'Fase de Grupos',
    'second round': 'Fase de Grupos',
    'second group stage': 'Fase de Grupos',
    'group stage play-off': 'Fase de Grupos',
    'final stage': 'Fase de Grupos'}
```

```python
matches ['Stage'] = matches['Round'].str.lower().map(fase_map)
```

```python
matches ['Stage'].value_counts()
```

```python
matches['Stage'].describe()
```

```python
def get_winner(row):
    if pd.notna(row['home_penalty']) and pd.notna(row['away_penalty']):
        if row['home_penalty'] > row['away_penalty']:
            return row['home_team']
        elif row['home_penalty'] < row['away_penalty']:
            return row['away_team']
        else:
            return 'Draw'
    elif row['home_score'] > row['away_score']:
        return row['home_team']
    elif row['home_score'] < row['away_score']:
        return row['away_team']
    else:
        return 'Draw'
```

```python
matches ['Winner'] = matches.apply(get_winner, axis=1  )
```

```python
matches [matches['Year'] == 2022][['home_team', 'away_team', 'home_score', 'away_score', 'home_penalty', 'away_penalty', 'Winner', 'Stage']].head() 
```

```python
def conta_cartao(x):
    if pd.isna(x):
        return 0
    else:
        return str(x).count('|') + 1
```

```python
matches['home_red_cards'] = matches['home_red_card'].apply(conta_cartao)
matches['away_red_cards'] = matches['away_red_card'].apply(conta_cartao)
matches['home_yellow_red_cards'] = matches['home_yellow_red_card'].apply(conta_cartao)
matches['away_yellow_red_cards'] = matches['away_yellow_red_card'].apply(conta_cartao)

```

```python
# Verifica
print(matches['home_red_cards'].value_counts())
print(matches['away_red_cards'].value_counts())
```

```python
print(matches['away_yellow_red_cards'].value_counts())
print(matches['home_yellow_red_cards'].value_counts())
```
